```python
import glob
import time
import pickle
import pandas as pd
import os
```

## Erstellung der List of Indices

Die tokenisierten Listen werden zu Index-Listen konvertiert. Dabei erhält jedes Token den Index, welches es in der mfws_Liste hat. So entsteht im Prinzip eine Liste aus Wort IDs. Wenn Token nicht in mfws_Liste ist, wird in der Tokenliste eines Texts die ID bzw der Index 10.000 gespeichert. 

```python
with open('pickles\\frequency_dist_counter.p', 'rb') as handle:
    frequency_dist = pickle.load(handle)
    
mc = frequency_dist.most_common(10000)
mc = [tupel[0] for tupel in mc]
mc
```

```python
grade = time.time()

for path in glob.glob("pickles\\tokenized_docs\*"):
    list_of_wordIDs = []
    
    
    print("Start",time.time() -grade)
    print(path)
    
    with open(path, 'rb') as handle:
        tokenized_docs = pickle.load(handle)
        
   
    for doc in tokenized_docs:
        #print("Neues Doc",time.time() -grade)
        l = [mc.index(token) if token in mc else 10000 for token in doc]
        list_of_wordIDs.append(l)
        
    with open("word_ids\\"+"word_ids_"+os.path.basename(path)+".p", 'wb') as handle:
        pickle.dump(list_of_wordIDs, handle)
        
```
