```python
import time
import numpy as np
import pandas as pd
import pickle
import glob
```

## Berechnen der Kookkurrenz Matrix

```python
co_occurence_matrix = np.zeros((10002,10002))

document_frequency_matrix= np.zeros((10002,10002))

padding_wordIDs = [10001,10001]

for path in glob.glob("word_ids\*"): 
    print(path)
    
    with open(path, 'rb') as file:
        list_of_wordIDs = pickle.load(file)
        
    start = time.time()

    
    for doc in list_of_wordIDs: 
        
        tupel_set = set()
        doc = padding_wordIDs + doc + padding_wordIDs
        
        for i, center_word in enumerate(doc[2:-2]):
            i=i+2
                        
            window = 1
            while window <= 2:

                co_occurence_matrix[doc[i-window], center_word] +=1
                co_occurence_matrix[doc[i+window], center_word] +=1
                
                tupel_set.add((doc[i-window], center_word))
                tupel_set.add((doc[i+window], center_word))

                
            
                window +=1  
                
        for tupel in tupel_set:
            document_frequency_matrix[tupel] += 1

    
    print("Ende ", time.time()-start)


with open("matrix\\"+"co_occ_matrix.p", 'wb') as handle:
    pickle.dump(co_occurence_matrix, handle)
    
    
with open("matrix\\"+"document_freq_matrix.p", 'wb') as handle:
    pickle.dump(document_frequency_matrix, handle)
        
    
        
```

## Create numpy as dataframe

```python
with open('pickles\\frequency_dist_counter.p', 'rb') as handle:
    frequency_dist = pickle.load(handle)
    
mfws_list = frequency_dist.most_common(10000)
mfws_list = [tupel[0] for tupel in mfws_list]



# Nur einmal ausführen, damit die mfws_liste noch another_token und padding_token enthält

# mfws_list.append("another_token")
# mfws_list.append("padding_token")


# Testen der Indices

# print(mfws_list[10000])
# print(mfws_list[10001])
# print(len(mfws_list))
```

#### Convert numpy matrix to pandas dataframe

```python
co_occurence_matrix_df = pd.DataFrame(co_occurence_matrix, index = mfws_list, columns = mfws_list)
#co_occurence_matrix_df


document_frequency_matrix_df = pd.DataFrame( document_frequency_matrix, index = mfws_list, columns = mfws_list)
#document_frequency_df
```

#### Save pandas dataframes as pickles

```python
with open("matrix\\"+"co_occ_matrix_as_pandas_df.p", 'wb') as handle:
    pickle.dump(co_occurence_matrix_df, handle)
    
    
with open("matrix\\"+"document_freq_matrix_as_pandas_df.p", 'wb') as handle:
    pickle.dump(document_frequency_matrix_df, handle)
```
