## Implementierung von tfidf


#### Methoden

```python

def tfLog(tf): 
    if tf > 0:
        return 1+np.log10(tf)
    else: 
        return 0
        
def idf(N,df_t):
    if df_t > 0:
        return np.log10(N/df_t)
    else:
        return 0
    
    
    
```

#### Anwendung

```python
tfidf_matrix = np.zeros((10002,10002))

start = time.time()
 
for i, row in enumerate(co_occurence_np_matrix):
    print("Row ---",i,"---", time.time()-start)

    for j,column in enumerate(co_occurence_np_matrix.T):
        
        
        #print("Stelle ",i,j, "Kook-Wert ", tfLog(co_occurence_matrix[i,j]), "ZugehÃ¶riger df(kook) Wert", document_frequency_matrix[i,j])
        
        tfidf_matrix[i,j] = tfLog(co_occurence_np_matrix[i,j]) * idf(N,document_freq_np_matrix[i,j])

```
