```python
from bs4 import BeautifulSoup
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk
import collections
import glob
from nltk import FreqDist
from collections import Counter
import pickle
import os
from io import StringIO
import pandas as pd
```

## Stopwords

```python
custom_stops = open("..\de.txt", "r", encoding="utf-8").read().splitlines()

stopset = stopwords.words('german')
stopset.extend(custom_stops)
stopset = set(stopset)
#print(stopset)
print(len(stopset))
```

## Tokenisieren der Wiki Dateien, Erstellen der Word Frequency Dist

Das ist die endgültige Version, die alle Dateien erstellt hat, welche auch zur weiteren Verarbeitung verwendet wurden. Dabei werden die tokenisierten Texte pro Ordner in Listen gespeichert und als Gesamtordner wieder gepickelt. Die Listen aus Listen werden im nächsten Schritt wieder ausgelesen, sodass nur eine Liste aus Tokens aus allen Texten eines Ordners entsteht. Die daraus entwickelte Frequency Distribution wird in einen Counter gespeichert, iterativ für alle Ordner. Sodass am Ende eine Frequency Distribution aller Wörter aus allen Wiki Dateien entsteht. Dies wird auch gepickelt

```python
import time


counter = Counter()

for path in glob.glob("..\\extracted\*"):
    grade = time.time()
    print(path)
    
    all_docs = []
   
    
    for wiki in glob.glob(path+"\*"):
        print(wiki)
    
        soup = BeautifulSoup(open(wiki, encoding="utf-8"), "html.parser")   
        
        
        for doc_tag in soup.find_all("doc"):
            
            doc = doc_tag.text
            
            words = word_tokenize(doc)
            
            words=[word.lower() for word in words if word.isalpha()]
    
            filtered_words = [word for word in words if word not in stopset]
        
            
            all_docs.append(filtered_words)
            
            
        
    with open("pickles\\"+"tokenized_documents_"+os.path.basename(path)+".p", 'wb') as handle:
        pickle.dump(all_docs, handle)
        
    
    flat_list = [item for sublist in all_docs for item in sublist]

    
    frequency_distribution = Counter(FreqDist(flat_list))
            
    counter = counter + frequency_distribution

    
with open("pickles\\"+"frequency_dist_counter.p", 'wb') as handle:
    pickle.dump(counter, handle)
    
        
    
```
